```python
people = pd.DataFrame(np.random.randn(5, 5),
                     columns='a b c d e'.split(),
                     index='Joe Steve Wes Jim Travis'.split())
people.iloc[2:3, [1, 2]] = np.nan
people
```

```python
mapping = {
    'a': 'red',
    'b': 'red',
    'c': 'blue',
    'd': 'blue',
    'e': 'red',
    'f': 'orange'
}
```

```python
by_column = people.groupby(mapping, axis=1)
```

```python
by_column.sum()
```

```python
rows = 100

foo = pd.DataFrame(np.random.randn(rows,5),
                   columns='a b c d e'.split())

foo['FY'] = np.random.randint(2012,2020,rows)

foo.head(10)
```

```python
def boolean(x):
    x = str(x)[-1:]
    return x in ('6', '2')
```

```python
bar = foo['FY'].apply(boolean).values
```

```python
bar
```

```python
g = foo.groupby(bar)
```

```python
g[['a', 'b']].mean()
```

```python
def foo2(arr):
    arr2 = arr ** 2
    return arr2.sum()/len(arr2)
```

```python
g[['a','b']].agg(foo2)
```

```python
g.describe()
```

```python
g.agg('mean')
```

```python
g[['a', 'b', 'c']].agg(['count', 'mean', 'max'])
```

```python
k = g[['a', 'b', 'c']]
```

```python
k.agg({
    'a': ['min', 'max', 'std'],
    'b': 'sum',
    'c': foo2
})
```

```python
k.agg({
    'a': 'min',
    'b': 'sum',
    'c': foo2
})
```

```python
def top(df, n=5, column='b'):
    return df.sort_values(by=column)[-n:]
```

```python
foo.groupby('FY').apply(top, n=3, column='a')
```
