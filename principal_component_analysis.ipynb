```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
```

```python
def dummy_data(n=100):
    return pd.DataFrame({
        'a': np.random.normal(0, 1, n),
        'b': np.random.normal(0, 1, n),
        'c': np.random.normal(0, 1, n),
        'd': np.arange(n)/n,
        'e': np.random.normal(0, 1, n),
        'f': np.random.normal(0, 1, n)
    })
```

```python
df = dummy_data(1000)
df.head()
```

```python
pca = PCA(n_components=3)
```

```python
data = df.values
columns = df.columns.values

print(columns)
print(data[:5, :3])
```

```python
X2D = pca.fit_transform(data)
```

```python
X2D[:5]
```

```python
t = X2D.T
t[:, :5]
```
